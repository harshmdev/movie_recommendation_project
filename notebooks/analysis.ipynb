```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("artifacts/data_transformation/tf_data.csv")
df.head()
```

```python
#Gross by director
#rating by director
#Distribution of rating,gross,runtime
#relation between runtime,rating,gross
#descriptive statistics of runtime, gross, rating
```

```python
temp_df=df.groupby("director")["gross"].mean().sort_values(ascending=False).reset_index()
```

```python
sns.barplot(data=temp_df.head(10),x="director",y="gross")
plt.title('Average Gross Earnings by Director')
plt.xlabel('Director Name')  # Change the x-axis title
plt.ylabel('Average Gross Earnings(100 Million)')

# Rotate the x-ticks for better readability
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python
temp_df=df.groupby("director")["rating"].sum().sort_values(ascending=False).reset_index()
```

```python
sns.barplot(data=temp_df.head(10),x="director",y="rating")
plt.title('Sum of Ratings by Director')
plt.xlabel('Director Name')  # Change the x-axis title
plt.ylabel('Sum of Ratings')

# Rotate the x-ticks for better readability
plt.xticks(rotation=45, ha='right') 
plt.show()
```
